```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

class ModifiedICSSDetector:
    """
    Modified Iterative Cumulative Sum of Squares (ICSS) algorithm
    for detecting structural breaks in variance of return series.
    
    Based on Inclan & Tiao (1994) with modifications for practical application.
    """
    
    def __init__(self, significance_level=0.05):
        """
        Initialize the ICSS detector.
        
        Parameters:
        -----------
        significance_level : float
            Significance level for the test (default: 0.05)
        """
        self.alpha = significance_level
        self.breaks = []
        self.test_statistics = []
        
    def _get_critical_value(self, n):
        """
        Get critical value based on sample size.
        Uses asymptotic critical values from Inclan & Tiao (1994).
        
        Parameters:
        -----------
        n : int
            Sample size
        """
        # Asymptotic critical values for different significance levels
        critical_values = {
            0.10: 1.22,
            0.05: 1.36,
            0.01: 1.63
        }
        return critical_values.get(self.alpha, 1.36)
    
    def _compute_cusum_statistic(self, returns):
        """
        Compute the ICSS test statistic (CUSUM of squares).
        
        Parameters:
        -----------
        returns : array-like
            Return series
            
        Returns:
        --------
        Dk : array
            Test statistic values
        k_max : int
            Location of maximum test statistic
        D_max : float
            Maximum test statistic value
        """
        n = len(returns)
        centered_returns = returns - np.mean(returns)
        squared_returns = centered_returns ** 2
        
        # Cumulative sum of squared returns
        C_k = np.cumsum(squared_returns)
        C_T = C_k[-1]
        
        # Compute test statistic for each k
        k_range = np.arange(1, n)
        D_k = np.sqrt(n / 2) * (C_k[:-1] / C_T - k_range / n)
        
        # Find maximum absolute value
        k_max = np.argmax(np.abs(D_k)) + 1  # +1 because we started from index 1
        D_max = np.abs(D_k[k_max - 1])
        
        return D_k, k_max, D_max
    
    def _detect_breaks_recursive(self, returns, dates, start_idx=0):
        """
        Recursively detect structural breaks.
        
        Parameters:
        -----------
        returns : array-like
            Return series
        dates : array-like
            Corresponding dates
        start_idx : int
            Starting index in the original series
        """
        n = len(returns)
        
        # Need minimum sample size
        if n < 30:
            return
        
        D_k, k_max, D_max = self._compute_cusum_statistic(returns)
        critical_value = self._get_critical_value(n)
        
        # If test statistic exceeds critical value, we found a break
        if D_max > critical_value:
            break_idx = start_idx + k_max
            self.breaks.append({
                'index': break_idx,
                'date': dates[k_max],
                'statistic': D_max,
                'critical_value': critical_value
            })
            
            # Recursively test sub-samples
            # Left sub-sample
            if k_max > 30:
                self._detect_breaks_recursive(
                    returns[:k_max], 
                    dates[:k_max], 
                    start_idx
                )
            
            # Right sub-sample
            if n - k_max > 30:
                self._detect_breaks_recursive(
                    returns[k_max:], 
                    dates[k_max:], 
                    start_idx + k_max
                )
    
    def detect_breaks(self, returns_series):
        """
        Main method to detect structural breaks in a return series.
        
        Parameters:
        -----------
        returns_series : pd.Series
            Pandas Series with DatetimeIndex containing log returns
            
        Returns:
        --------
        breaks_df : pd.DataFrame
            DataFrame containing detected break points
        """
        # Reset breaks from previous runs
        self.breaks = []
        
        # Remove NaN values
        returns_clean = returns_series.dropna()
        returns = returns_clean.values
        dates = returns_clean.index
        
        # Detect breaks recursively
        self._detect_breaks_recursive(returns, dates)
        
        # Sort breaks by index
        self.breaks.sort(key=lambda x: x['index'])
        
        # Create DataFrame
        if self.breaks:
            breaks_df = pd.DataFrame(self.breaks)
            return breaks_df
        else:
            return pd.DataFrame(columns=['index', 'date', 'statistic', 'critical_value'])
    
    def plot_results(self, returns_series, figsize=(14, 10)):
        """
        Visualize the return series with detected structural breaks.
        
        Parameters:
        -----------
        returns_series : pd.Series
            Pandas Series with DatetimeIndex containing log returns
        figsize : tuple
            Figure size (width, height)
        """
        returns_clean = returns_series.dropna()
        
        fig, axes = plt.subplots(3, 1, figsize=figsize)
        
        # Plot 1: Returns with break points
        axes[0].plot(returns_clean.index, returns_clean.values, 
                     linewidth=0.8, alpha=0.7, color='blue')
        axes[0].set_title('Log Returns with Detected Structural Breaks', 
                         fontsize=12, fontweight='bold')
        axes[0].set_ylabel('Log Returns')
        axes[0].grid(True, alpha=0.3)
        
        # Mark break points
        for break_point in self.breaks:
            axes[0].axvline(x=break_point['date'], color='red', 
                          linestyle='--', linewidth=1.5, alpha=0.7)
        
        # Plot 2: Rolling volatility
        rolling_vol = returns_clean.rolling(window=30).std() * np.sqrt(252)
        axes[1].plot(rolling_vol.index, rolling_vol.values, 
                    linewidth=1.2, color='green')
        axes[1].set_title('Rolling Volatility (30-day, annualized)', 
                         fontsize=12, fontweight='bold')
        axes[1].set_ylabel('Volatility')
        axes[1].grid(True, alpha=0.3)
        
        # Mark break points
        for break_point in self.breaks:
            axes[1].axvline(x=break_point['date'], color='red', 
                          linestyle='--', linewidth=1.5, alpha=0.7)
        
        # Plot 3: Variance regimes
        variance_regimes = self._compute_variance_regimes(returns_clean)
        axes[2].plot(variance_regimes.index, variance_regimes.values, 
                    linewidth=1.5, color='purple', drawstyle='steps-post')
        axes[2].set_title('Variance Regimes (between break points)', 
                         fontsize=12, fontweight='bold')
        axes[2].set_ylabel('Variance')
        axes[2].set_xlabel('Date')
        axes[2].grid(True, alpha=0.3)
        
        # Mark break points
        for break_point in self.breaks:
            axes[2].axvline(x=break_point['date'], color='red', 
                          linestyle='--', linewidth=1.5, alpha=0.7,
                          label='Break Point' if break_point == self.breaks[0] else '')
        
        if self.breaks:
            axes[2].legend()
        
        plt.tight_layout()
        plt.show()
        
        return fig
    
    def _compute_variance_regimes(self, returns_series):
        """
        Compute variance for each regime between break points.
        
        Parameters:
        -----------
        returns_series : pd.Series
            Clean return series
            
        Returns:
        --------
        variance_series : pd.Series
            Series with variance value for each regime
        """
        if not self.breaks:
            # No breaks, single regime
            return pd.Series(np.var(returns_series), 
                           index=returns_series.index)
        
        # Create regime indicators
        break_indices = [0] + [b['index'] for b in self.breaks] + [len(returns_series)]
        
        variance_values = np.zeros(len(returns_series))
        
        for i in range(len(break_indices) - 1):
            start_idx = break_indices[i]
            end_idx = break_indices[i + 1]
            regime_returns = returns_series.iloc[start_idx:end_idx]
            regime_var = np.var(regime_returns)
            variance_values[start_idx:end_idx] = regime_var
        
        return pd.Series(variance_values, index=returns_series.index)
    
    def get_summary(self):
        """
        Get a summary of detected breaks.
        
        Returns:
        --------
        summary : dict
            Dictionary containing summary statistics
        """
        if not self.breaks:
            return {
                'num_breaks': 0,
                'message': 'No structural breaks detected'
            }
        
        return {
            'num_breaks': len(self.breaks),
            'break_dates': [b['date'] for b in self.breaks],
            'test_statistics': [b['statistic'] for b in self.breaks],
            'significance_level': self.alpha
        }
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
from scipy.stats import norm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


class GARCHWithBreaks:
    """
    GARCH(1,1) model with structural breaks incorporated as regime-specific indicators.
    
    Variance equation: σ²_t = ω + α·ε²_{t-1} + β·σ²_{t-1} + Σ(δ_i · I_i,t)
    
    where I_i,t = 1 if observation t is in regime i, 0 otherwise.
    """
    
    def __init__(self, break_dates=None):
        """
        Initialize GARCH model with structural breaks.
        
        Parameters:
        -----------
        break_dates : list or None
            List of break point dates (pd.Timestamp objects)
        """
        self.break_dates = break_dates if break_dates is not None else []
        self.params = None
        self.convergence = None
        self.log_likelihood = None
        self.conditional_volatility = None
        self.standardized_residuals = None
        
    def _create_regime_indicators(self, dates):
        """
        Create regime-specific indicator variables.
        
        Parameters:
        -----------
        dates : pd.DatetimeIndex
            Date index of the return series
            
        Returns:
        --------
        indicators : np.ndarray
            Array of shape (T, n_regimes) with regime indicators
        """
        T = len(dates)
        n_regimes = len(self.break_dates) + 1
        indicators = np.zeros((T, n_regimes))
        
        # Sort break dates
        sorted_breaks = sorted(self.break_dates)
        
        # Create regime boundaries
        regime_starts = [dates[0]] + sorted_breaks
        regime_ends = sorted_breaks + [dates[-1]]
        
        for i in range(n_regimes):
            mask = (dates >= regime_starts[i]) & (dates <= regime_ends[i])
            indicators[mask, i] = 1
            
        return indicators
    
    def _garch_likelihood(self, params, returns, indicators):
        """
        Compute negative log-likelihood for GARCH(1,1) with structural breaks.
        
        Parameters:
        -----------
        params : array
            Parameters [mu, omega, alpha, beta, delta_1, ..., delta_k]
        returns : array
            Return series
        indicators : np.ndarray
            Regime indicator matrix
            
        Returns:
        --------
        neg_log_likelihood : float
            Negative log-likelihood value
        """
        T = len(returns)
        n_regimes = indicators.shape[1]
        
        # Unpack parameters
        mu = params[0]
        omega = params[1]
        alpha = params[2]
        beta = params[3]
        deltas = params[4:4+n_regimes]  # Regime-specific shifts
        
        # Parameter constraints (return large value if violated)
        if omega <= 0 or alpha < 0 or beta < 0 or alpha + beta >= 1:
            return 1e10
        
        # Initialize
        epsilon = returns - mu
        sigma2 = np.zeros(T)
        sigma2[0] = np.var(returns)  # Initialize with sample variance
        
        # Compute conditional variances
        for t in range(1, T):
            # Base GARCH variance
            base_variance = omega + alpha * epsilon[t-1]**2 + beta * sigma2[t-1]
            
            # Add regime-specific shifts
            regime_shift = np.sum(deltas * indicators[t, :])
            
            sigma2[t] = base_variance + regime_shift
            
            # Ensure variance is positive
            if sigma2[t] <= 0:
                return 1e10
        
        # Compute log-likelihood
        log_likelihood = -0.5 * np.sum(
            np.log(2 * np.pi) + np.log(sigma2) + (epsilon**2 / sigma2)
        )
        
        return -log_likelihood  # Return negative for minimization
    
    def fit(self, returns_series, method='L-BFGS-B', display=True):
        """
        Fit GARCH model with structural breaks using maximum likelihood.
        
        Parameters:
        -----------
        returns_series : pd.Series
            Time series of returns with DatetimeIndex
        method : str
            Optimization method (default: 'L-BFGS-B')
        display : bool
            Whether to display estimation results
            
        Returns:
        --------
        self : GARCHWithBreaks
            Fitted model instance
        """
        # Clean data
        returns_clean = returns_series.dropna()
        returns = returns_clean.values
        dates = returns_clean.index
        
        # Create regime indicators
        indicators = self._create_regime_indicators(dates)
        n_regimes = indicators.shape[1]
        
        # Initial parameter values
        # [mu, omega, alpha, beta, delta_1, ..., delta_k]
        mu_init = np.mean(returns)
        var_init = np.var(returns)
        omega_init = var_init * 0.01
        alpha_init = 0.08
        beta_init = 0.90
        delta_init = [0.0] * n_regimes  # Start with no regime shifts
        
        initial_params = np.array([mu_init, omega_init, alpha_init, beta_init] + delta_init)
        
        # Parameter bounds
        bounds = [
            (None, None),          # mu (no bounds)
            (1e-6, None),          # omega > 0
            (0, 1),                # 0 <= alpha < 1
            (0, 1),                # 0 <= beta < 1
        ] + [(None, None)] * n_regimes  # deltas (no bounds)
        
        # Optimize
        if display:
            print("Estimating GARCH(1,1) with structural breaks...")
            print(f"Number of regimes: {n_regimes}")
            print(f"Sample size: {len(returns)}")
        
        result = minimize(
            self._garch_likelihood,
            initial_params,
            args=(returns, indicators),
            method=method,
            bounds=bounds,
            options={'maxiter': 1000}
        )
        
        self.convergence = result
        self.params = result.x
        self.log_likelihood = -result.fun
        
        # Compute conditional volatility with fitted parameters
        self._compute_conditional_volatility(returns, indicators)
        
        if display:
            self.print_results(returns_series)
        
        return self
    
    def _compute_conditional_volatility(self, returns, indicators):
        """
        Compute conditional volatility using fitted parameters.
        """
        mu = self.params[0]
        omega = self.params[1]
        alpha = self.params[2]
        beta = self.params[3]
        deltas = self.params[4:]
        
        T = len(returns)
        epsilon = returns - mu
        sigma2 = np.zeros(T)
        sigma2[0] = np.var(returns)
        
        for t in range(1, T):
            base_variance = omega + alpha * epsilon[t-1]**2 + beta * sigma2[t-1]
            regime_shift = np.sum(deltas * indicators[t, :])
            sigma2[t] = base_variance + regime_shift
        
        self.conditional_volatility = np.sqrt(sigma2)
        self.standardized_residuals = epsilon / self.conditional_volatility
    
    def print_results(self, returns_series):
        """
        Print estimation results.
        """
        n_regimes = len(self.break_dates) + 1
        
        print("\n" + "="*70)
        print("GARCH(1,1) MODEL WITH STRUCTURAL BREAKS - ESTIMATION RESULTS")
        print("="*70)
        
        print(f"\nLog-Likelihood: {self.log_likelihood:.4f}")
        print(f"Convergence: {self.convergence.success}")
        
        print("\n" + "-"*70)
        print("GARCH PARAMETERS:")
        print("-"*70)
        print(f"{'Parameter':<15} {'Estimate':<15} {'Description'}")
        print("-"*70)
        print(f"{'mu':<15} {self.params[0]:>14.6f}  Mean return")
        print(f"{'omega':<15} {self.params[1]:>14.6e}  Constant term")
        print(f"{'alpha':<15} {self.params[2]:>14.6f}  ARCH effect")
        print(f"{'beta':<15} {self.params[3]:>14.6f}  GARCH effect")
        print(f"{'alpha+beta':<15} {self.params[2]+self.params[3]:>14.6f}  Persistence")
        
        print("\n" + "-"*70)
        print("REGIME-SPECIFIC VARIANCE SHIFTS (δ_i):")
        print("-"*70)
        print(f"{'Regime':<10} {'Period':<30} {'Delta (δ)':<15}")
        print("-"*70)
        
        sorted_breaks = sorted(self.break_dates)
        dates = returns_series.dropna().index
        
        # Regime 0 (before first break)
        period_start = dates[0].strftime('%Y-%m-%d')
        if len(sorted_breaks) > 0:
            period_end = sorted_breaks[0].strftime('%Y-%m-%d')
        else:
            period_end = dates[-1].strftime('%Y-%m-%d')
        print(f"{'0 (base)':<10} {period_start} to {period_end:<8} {self.params[4]:>14.6e}")
        
        # Subsequent regimes
        for i in range(1, n_regimes):
            period_start = sorted_breaks[i-1].strftime('%Y-%m-%d')
            if i < len(sorted_breaks):
                period_end = sorted_breaks[i].strftime('%Y-%m-%d')
            else:
                period_end = dates[-1].strftime('%Y-%m-%d')
            print(f"{i:<10} {period_start} to {period_end:<8} {self.params[4+i]:>14.6e}")
        
        print("="*70 + "\n")
    
    def plot_volatility(self, returns_series, figsize=(14, 10)):
        """
        Plot conditional volatility with regime breaks.
        
        Parameters:
        -----------
        returns_series : pd.Series
            Original return series
        figsize : tuple
            Figure size
        """
        if self.conditional_volatility is None:
            raise ValueError("Model must be fitted before plotting")
        
        returns_clean = returns_series.dropna()
        dates = returns_clean.index
        
        fig, axes = plt.subplots(3, 1, figsize=figsize)
        
        # Plot 1: Returns
        axes[0].plot(dates, returns_clean.values, linewidth=0.6, alpha=0.7, color='blue')
        axes[0].set_title('Log Returns', fontsize=12, fontweight='bold')
        axes[0].set_ylabel('Returns')
        axes[0].grid(True, alpha=0.3)
        
        for break_date in self.break_dates:
            axes[0].axvline(x=break_date, color='red', linestyle='--', 
                          linewidth=1.5, alpha=0.6)
        
        # Plot 2: Conditional Volatility
        annualized_vol = self.conditional_volatility * np.sqrt(252)
        axes[1].plot(dates, annualized_vol, linewidth=1.2, color='green', 
                    label='Conditional Volatility')
        axes[1].set_title('GARCH Conditional Volatility (Annualized)', 
                         fontsize=12, fontweight='bold')
        axes[1].set_ylabel('Volatility')
        axes[1].grid(True, alpha=0.3)
        axes[1].legend()
        
        for break_date in self.break_dates:
            axes[1].axvline(x=break_date, color='red', linestyle='--', 
                          linewidth=1.5, alpha=0.6, label='Break' if break_date == self.break_dates[0] else '')
        
        # Plot 3: Standardized Residuals
        axes[2].plot(dates, self.standardized_residuals, linewidth=0.6, 
                    alpha=0.7, color='purple')
        axes[2].axhline(y=0, color='black', linestyle='-', linewidth=0.8)
        axes[2].axhline(y=2, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)
        axes[2].axhline(y=-2, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)
        axes[2].set_title('Standardized Residuals', fontsize=12, fontweight='bold')
        axes[2].set_ylabel('Std. Residuals')
        axes[2].set_xlabel('Date')
        axes[2].grid(True, alpha=0.3)
        
        for break_date in self.break_dates:
            axes[2].axvline(x=break_date, color='red', linestyle='--', 
                          linewidth=1.5, alpha=0.6)
        
        plt.tight_layout()
        plt.show()
        
        return fig
    
    def get_regime_volatilities(self, returns_series):
        """
        Calculate average volatility for each regime.
        
        Returns:
        --------
        regime_stats : pd.DataFrame
            DataFrame with regime statistics
        """
        if self.conditional_volatility is None:
            raise ValueError("Model must be fitted first")
        
        returns_clean = returns_series.dropna()
        dates = returns_clean.index
        indicators = self._create_regime_indicators(dates)
        n_regimes = indicators.shape[1]
        
        regime_stats = []
        sorted_breaks = sorted(self.break_dates)
        
        for i in range(n_regimes):
            regime_mask = indicators[:, i] == 1
            regime_vol = self.conditional_volatility[regime_mask]
            
            # Get regime period
            if i == 0:
                period_start = dates[0]
                period_end = sorted_breaks[0] if sorted_breaks else dates[-1]
            elif i == n_regimes - 1:
                period_start = sorted_breaks[i-1]
                period_end = dates[-1]
            else:
                period_start = sorted_breaks[i-1]
                period_end = sorted_breaks[i]
            
            regime_stats.append({
                'Regime': i,
                'Start': period_start,
                'End': period_end,
                'Avg_Volatility': np.mean(regime_vol) * np.sqrt(252),
                'Min_Volatility': np.min(regime_vol) * np.sqrt(252),
                'Max_Volatility': np.max(regime_vol) * np.sqrt(252),
                'Delta': self.params[4+i]
            })
        
        return pd.DataFrame(regime_stats)

```
